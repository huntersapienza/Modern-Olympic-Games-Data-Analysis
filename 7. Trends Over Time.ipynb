To start, we import the necessary packages and upload the cleaned data from the introduction section.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('cleaned_data.csv')
```

Given that we will be working with athlete age, height, and weight, we will also upload the original data as well, to avoid the distribution bias that goes along with filling all the null values with the median for each characteristic.

```python
unclean_data = pd.read_csv('athlete_events.csv')
```

To prepare for visualizations, we group the original data by year and compute the mean for each continuous feature. The first five rows of the results are shown in the dataframe below, and throughout the next chunk of this notebook, we will visualize the entire dataframe as barplots over time.

```python
mean_unclean = unclean_data.groupby('Year').mean()
```

```python
mean_unclean.reset_index(inplace=True)
```

```python
mean_unclean.head()
```

## How have athlete age, height, and weight changed over time?

### Age Over Time

Below, we visualize average athlete age over time as a barplot, boxplot, and pointplot.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.barplot(mean_unclean['Year'],mean_unclean['Age'],ci=False)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Age',size=25)
plt.title('Average Athlete Age Over Time',fontsize=30)
plt.show()
```

Based on the plot of average age over time above, we can see that the average fluctuated a bit, reaching highs in 1900, 1920, 1932, and 1948, before reaching a low in 1980 and finally becoming relatively stable in the 21st century. The relative instability in early athlete age data may be caused by two factors: (1) the presence of significant missing data, and (2) early events with older competitors such as roque, art competitions, and jeu de paume. The low in 1980 may come from the presence of many extraordinarily young athletes prior to the age limit of 15 enforced by the IOC in 1981, and then raised to 16 in 1997. In the past couple decades, however, the average age has settle around 25 or 26 years old.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.boxplot('Year','Age',data=unclean_data)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Age',size=25)
plt.title('Distribution of Athlete Age Over Time',fontsize=30)
plt.show()
```

The boxplots above illustrate the age distribution by quartiles for age at each year of the Olympic Games. In the early years of the games, we see that the box plots have much greater ranges than in recent years, supporting the theory that age is stabilizing and the window is narrowing for individual's to compete in the games. We see far fewer young outliers than old, given that the average athlete age is already low to begin with, but we see the oldest athletes in the late 1920s and throughout the 1930s, whereas there appear to be much younger "oldest athletes" at the recent games. Some young outliers participated in the 1980s and 90s, prior to enforcement of the 16 year age mimimum.

```python
plt.figure(figsize=(30, 15))
sns.pointplot('Year', 'Age', data=unclean_data)
plt.title('Variation of Age for Athletes over time',size=30)
plt.xlabel('Year',size=25)
plt.ylabel('Age',size=25)
plt.xticks(rotation=90,size=20)
plt.show()
```

In the point plot above, we see few additional trends to those mentioned above. The most significant to observe is the narrowing error bars over time, with greater variation in age from 1896 to 1948, but very small confidence intervals in recent years as this age range for athletes continues to narrow.

### Height Over Time

Below, we visualize average athlete height over time as a barplot, boxplot, and pointplot.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.barplot(mean_unclean['Year'],mean_unclean['Height'],ci=False)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Height',size=25)
plt.title('Average Athlete Height Over Time',fontsize=30)
plt.show()
```

Over the year, average athlete height has remained very consistent, with just some minor variation. After the split of the summer and winter games in 1992, we see an oscillating pattern that indicates slight variation in the average height of summer and winter athletes, although based on our results in notebook 6. Seasonal Athletes, we determined this difference is not statistically significant.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.boxplot('Year','Height',data=unclean_data)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Height',size=25)
plt.title('Distribution of Athlete Height Over Time',fontsize=30)
plt.show()
```

The boxplots above illustrate the height distribution by quartiles at each year of the Olympic Games. Opposite our results for the age distributions, it appears that variation in height is increasing over time with seemingly more outliers on either side of the plots' whiskers and a greater range between the ends of our plots. This phenomenon may be the result of increasing differentiation and specialization within each specific sport at the Olympic Games, as well as an increasing number of sports for which this occurs as the program at the games grows. More colloquially, the characteristics that best serve a gymnasts do not best serve a basketball player, and vice versa. Thus, the range of athlete body types and characteristics appears to be diverging, although the center average remains relatively unchanged.

```python
plt.figure(figsize=(30, 15))
sns.pointplot('Year', 'Height', data=unclean_data)
plt.title('Variation of Height for Athletes over time',size=30)
plt.xlabel('Year',size=25)
plt.ylabel('Height',size=25)
plt.xticks(rotation=90,size=20)
plt.show()
```

The pointplot above, like the one for age over time, shows that the error in our average for height is decreasing over time, as we gather greater amounts of data (with an increasing number of athletes at the games) and thus the confidence intervals for our statistics shrink. Additionally, the oscillating trend in height between summer and winter games is more evident here, with winter athlete height seeminingly less than that for summer athletes.

### Weight Over Time

Below, we visualize average athlete weight over time as a barplot, boxplot, and pointplot.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.barplot(mean_unclean['Year'],mean_unclean['Weight'],ci=False)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Weight',size=25)
plt.title('Average Athlete Weight Over Time',fontsize=30)
plt.show()
```

Similar to the barplot for age over time, we see general instability in the average weight during the early modern Olympic Games. However, this appears to stabilize and remain relatively constant starting in 1960, with some slight variation between summer and winter games beginning in 2002, but again, based on the previous notebook, these differences were not statistically significant.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.boxplot('Year','Weight',data=unclean_data)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Weight',size=25)
plt.title('Distribution of Athlete Weight Over Time',fontsize=30)
plt.show()
```

The boxplots above illustrate the weight distribution by quartiles at each year of the Olympic Games. Opposite our results for the age distributions but very similar to those for height, it appears that variation in weight is increasing over time with seemingly more outliers on the upper side of the plots' whiskers and a greater range between the ends of our plots. The same factors that we identified in our analysis of the height boxplots above likely cause this phenomenon here as well, with greater differentiation occurring between athletes of different sports. However, the median value here appears to be very consistent, varying no more than ever so slightly following the 1956 games. In addition to the slight oscillatory pattern we noted in the last chart, significantly more outliers exist for the winter games than the summer following their split in 1992, with weights sometimes exceeding 200 kg in the summer games, while the greatest outlier in the winter games appears to be approximately 125 kg.

```python
plt.figure(figsize=(30, 15))
sns.pointplot('Year', 'Weight', data=unclean_data)
plt.title('Variation of Weight for Athletes over time',size=30)
plt.xlabel('Year',size=25)
plt.ylabel('Weight',size=25)
plt.xticks(rotation=90,size=20)
plt.show()
```

Between the three pointplots we have observed thus far, this one for weight demonstrates the largest error bars for all points throughout history. Although these confidence intervals narrow over time, just like for age and height, they seem to be much larger in recent years than their counterparts for the other characteristics. This is likely due to the significant upper outliers that we noted in the last chart, causing decreased confidence in our computed averages and thus increasing our error. 

## How has country participation changed over time?

As noted in previous notebooks, the size of the Olympic Games has grown tremendously since it's origin in 1896. Throughout the rest of the this section, we will explore the ways in which the size of the games has changed, beginning with the number of countries attending each games. Below, with create a new dataframe, grouped by the year and showing the numberer of unique NOC codes used each year.

```python
unique_countries = pd.DataFrame(unclean_data.groupby('Year')['NOC'].nunique())
unique_countries.reset_index(inplace=True)
unique_countries.columns = ['Year','Count']
```

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.barplot(unique_countries['Year'],unique_countries['Count'],ci=False)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Number of Countries',size=25)
plt.title('Country Participation in the Olympics Over Time',fontsize=30)
plt.show()
```

The chart above illustrates the results of our new dataframe, with the number of countries (based on NOC codes used) per year. As we can see, the number starts out very low in 1896, with approximately 12 countries at the games, but jumps to over 25 in the 1900 Paris games, before returning to a lower number and steadily increasing until around 1948, when, after World War II, the size of the games truly takes off! With the exception of drops in 1976 and 1980 due to political boycotts on the games, the number continues to rise until settling just over 200 NOC codes. With the separation of the summer and winter games in 1992, we can see a drastic difference in the number of countries at each. Thus, below, we will separate our country participation data into two separate dataframes for each season before proceeding with our analysis.

```python
summer_unique_countries = pd.DataFrame(unclean_data[unclean_data.Season=='Summer'].groupby('Year')['NOC'].nunique())
winter_unique_countries = pd.DataFrame(unclean_data[unclean_data.Season=='Winter'].groupby('Year')['NOC'].nunique())
```

```python
summer_unique_countries.reset_index(inplace=True)
summer_unique_countries.columns = ['Year','Count']
winter_unique_countries.reset_index(inplace=True)
winter_unique_countries.columns = ['Year','Count']
```

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.barplot(summer_unique_countries['Year'],summer_unique_countries['Count'],ci=False)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Number of Countries',size=25)
plt.title('Country Participation in the Summer Olympics Over Time',fontsize=30)
plt.show()
```

Above, in the chart representing the number of countries at each Summer Olympic Games, we see very similar numbers and patterns to that in our first chart, but without the oscillation after 1992 due to the lower attendance of the winter games.

```python
f, ax = plt.subplots(figsize = (30, 15))
sns.barplot(winter_unique_countries['Year'],winter_unique_countries['Count'],ci=False)
plt.xticks(rotation = 90,size=20)
plt.xlabel('Year',size=25)
plt.ylabel('Number of Countries',size=25)
plt.title('Country Participation in the Winter Olympics Over Time',fontsize=30)
plt.show()
```

The chart above represents the number of countries attending each Winter Olympic Games. We see that from its start in 1924, the winter games grow relatively steadily until 1980, after which they truly take off and increase in size at a much quicker rate. With over 80 countries in attendance in 2014, the winter games are becoming a more comparable counterpart to the summer games, and going forward I am sure we will see a continuation of this trend as the winter games build their prominence in the athletic community.

```python
participation_data.head()
```

Let's prepare for the map visualizations we'll create by making a method that takes in four inputs - the dataframe, a specific criteria, title keyword, and colorscale - and outputs a world choropleth map.

```python
def create_worldmap(data, criteria, title, color):
    trace = go.Choropleth(
                locations = data['region'],
                locationmode='country names',
                z = data[criteria],
                text = data[criteria],
                autocolorscale =False,
                reversescale = False,
                colorscale = color,
                marker = dict(
                    line = dict(
                        color = 'rgb(0,0,0)',
                        width = 0.5)
                ),
                colorbar = dict(
                    title = 'Number of Games',
                    tickprefix = '')
            )

    data = [trace]
    layout = go.Layout(
        title = 'Number of {} Olympic Games Participated in by Each Country'.format(title),
        geo = dict(
            showframe = True,
            showlakes = False,
            showcoastlines = True,
            projection = dict(
                type = 'natural earth'
            )
        )
    )

    fig = dict( data=data, layout=layout )
    iplot(fig)
```

## How has the overall size of the Olympic Games changed over time?

For our final analysis of Olympic Games size, we will explore how the number of athletes has changed over time. 

```python
data.head()
```

```python
host_cities = data.copy()[['Games','Year','Season','City']]
```

```python
host_cities.head()
```

```python
host_cities = host_cities.groupby(['Year','City','Season']).count()
```

```python
host_cities.reset_index(inplace=True)
```

Below, our new dataframe displays the number of athletes at each Olympic Games by year, with the host city and season specified as well.

```python
host_cities
```

Next, we separate this data by season for graphing purposes, to better visualize the change in size over time, given the significant differences between the summer and winter games.

```python
summer_cities = host_cities[host_cities['Season']=='Summer']
winter_cities = host_cities[host_cities['Season']=='Winter']
```

```python
plt.figure(figsize=(30,15))
sns.lineplot(x="Year", y="Games", hue="Season",data=host_cities,markers=True)
plt.title('Size of Olympic Games Over Time',size=30)
plt.xlabel('Year',size=25)
plt.ylabel('Total Athletes',size=25)
plt.xticks(rotation=90,size=20)
plt.legend()
plt.show()
```

In our graph of athlete participation over time, we see many similar trends to those visualized before, as well as the emergence of some new insights. Since 1896 and 1924, the origin points of the summer and winter games, we see both rising in athlete attendance ever since. However, in the summer games plot, we see unusual drops in attendance in 1932, 1956, and 1980. While we now that the drop in 1980 was the result of the boycott on the Soviet Union initiated by the United States, the other drops in 1932 and 1956 have less obvious causes. The winter games do not have such extreme drops in attendance, with only minor dips and rises outside the ordinary rising trend. Overall, we see a rise in the number of athletes at both the summer and winter Olympic Games over the past 124 years, a pattern that will likely continue for the winter games, while the summer games appears to be plateauing in recent years.
