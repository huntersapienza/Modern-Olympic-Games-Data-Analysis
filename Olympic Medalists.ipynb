## Do characteristics of Olympic medalists differ significantly from non-medalists?

```python
from scipy import stats
print('Normality of medalist age distribution: ', stats.normaltest(nonnull_medalist_age.Age))
print('Normality of medalist height distribution: ', stats.normaltest(nonnull_medalist_height.Height))
print('Normality of medalist weight distribution: ', stats.normaltest(nonnull_medalist_weight.Weight))
```

```python
def get_sample(data, n):
    sample = []
    while len(sample) != n:
        x = np.random.choice(data)
        sample.append(x)
    
    return sample

def get_sample_mean(sample):
    return sum(sample) / len(sample)

def create_sample_distribution(data, dist_size=100, n=30):
    sample_dist = []
    while len(sample_dist) != dist_size:
        sample = get_sample(data, n)
        sample_mean = get_sample_mean(sample)
        sample_dist.append(sample_mean)
    
    return sample_dist
```

```python
nonmedalist_data = data[data.Medal == 'None']
```

```python
norm_medalist_weight = create_sample_distribution(nonnull_medalist_weight.Weight,5000,500)
norm_medalist_height = create_sample_distribution(nonnull_medalist_height.Height,5000,500)
norm_medalist_age = create_sample_distribution(nonnull_medalist_age.Age,5000,500)
```

```python
norm_nonmedalist_weight = create_sample_distribution(nonnull_nonmedalist_weight.Weight,5000,500)
norm_nonmedalist_height = create_sample_distribution(nonnull_nonmedalist_height.Height,5000,500)
norm_nonmedalist_age = create_sample_distribution(nonnull_nonmedalist_age.Age,5000,500)
```

### Normalized Age

```python
plt.figure(figsize=(20, 10))
sns.distplot(norm_medalist_age, label='Normalized Medalist Age', color='gold')
sns.distplot(norm_nonmedalist_age, label='Normalized Non-Medalist Age', color='gray')
plt.title('Normalized Distribution of Age', fontsize=16)
plt.xlabel('Age', fontsize=16)
plt.legend()
plt.show()
```

```python
print('Normality of normalized medalist age distribution: ', stats.normaltest(norm_medalist_age))
print('Normality of normalized non-medalist age distribution: ', stats.normaltest(norm_nonmedalist_age))
```

```python
x = ['Medalists', 'Non-Medalists']
y = [np.array(norm_medalist_age).mean(),np.array(norm_nonmedalist_age).mean()]
plt.bar(x,y,color=('gold','gray'))
plt.ylabel('Age')
plt.xlabel('Athletes')
plt.title('Average Age by Medal Status')
```

```python
print('Average Medalist Age',np.array(norm_medalist_age).mean())
print('Average Non-Medalist Age',np.array(norm_nonmedalist_age).mean())
print('Difference in Means',np.array(norm_medalist_age).mean()-np.array(norm_nonmedalist_age).mean())
```

```python
def Cohen_d(group1, group2):
    diff = group1.mean() - group2.mean()

    n1, n2 = len(group1), len(group2)
    var1 = group1.var()
    var2 = group2.var()

    pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
    
    d = diff / np.sqrt(pooled_var)
    
    return d
```

```python
Cohen_d(np.array(norm_medalist_age),np.array(norm_nonmedalist_age))
```

```python
np.var(norm_medalist_age) == np.var(norm_nonmedalist_age)
```

```python
stats.ttest_ind(norm_medalist_age, norm_nonmedalist_age, equal_var=False)
```
